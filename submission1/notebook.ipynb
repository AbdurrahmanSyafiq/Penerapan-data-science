# Data Preparation / Preprocessing

drop data NaN

```python
df = df.dropna()
df.head()
```

Mengubah tipe data attrition menjadi object

```python
def attrition(x):
    if x == 0:
        return 'Stay'
    else:
        return 'Out'

df['Attrition'] = df['Attrition'].apply(attrition)
```

Analisis attrition pada monthlyIncome

```python
mi_df = df.groupby(by=["Attrition"]).agg({
    "MonthlyIncome": "mean"
}).sort_values(by=["Attrition"], ascending=True)
mi_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# delivery_merge['delivery_status'] = pd.Categorical(delivery_merge['delivery_status'], ["Tepat waktu", "Terlambat"])
plt.figure(figsize=(10, 5))
colors_ = ["#D3D3D3", "#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(
    y="MonthlyIncome",
    x="Attrition",
    data=mi_df.sort_values(by="Attrition", ascending=False),
    palette=colors_
)
plt.title("Number of Attrition by MonthlyIncome", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

Membuat dataframe baru dengan hanya mengambil data karyawan yang telah out

```python
attrition_df = df[df["Attrition"] == 'Out']
attrition_df
```

Analisis jumlah karyawan berdasarkan salaryHike

```python
# Membuat kolom baru 'SalaryHike' berdasarkan 'PercentSalaryHike'
def categorize_salary_hike(x):
    if x <= 15:
        return 'Low'
    elif 16 <= x <= 20:
        return 'Medium'
    else:
        return 'High'

attrition_df['SalaryHike'] = attrition_df['PercentSalaryHike'].apply(categorize_salary_hike)
```

```python
sh_df = attrition_df.groupby(by=["Attrition", "SalaryHike"]).agg({
    "EmployeeId": "count",
}).sort_values(by=["SalaryHike"], ascending=True)
sh_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index untuk mendapatkan kolom kembali dari multi-index
sh_df = sh_df.reset_index()

# Membuat bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='SalaryHike', y='EmployeeId', hue='SalaryHike', data=sh_df)

plt.title('Number of Employees by SalaryHike')
plt.ylabel('Number of Employees')
plt.xlabel('Salary Hike')
plt.show()
```

Analisis jumlah karyawan berdasarkan salaryHike dan OverTime

```python
so_df = attrition_df.groupby(by=["Attrition", "SalaryHike", "OverTime"]).agg({
    "EmployeeId": "count",
}).sort_values(by=["SalaryHike"], ascending=True)
so_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index untuk mendapatkan kolom kembali dari multi-index
so_df = so_df.reset_index()

# Membuat bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='OverTime', y='EmployeeId', hue='SalaryHike', data=so_df)

plt.title('Number of Employees by Overtime and SalaryHike')
plt.ylabel('Number of Employees')
plt.xlabel('Overtime')
plt.legend(title='SalaryHike')
plt.show()
```

Analisis jumlah karyawan berdasarkan jobInvolvement dan SalaryHike

```python
# Membuat kolom baru 'SalaryHike' berdasarkan 'PercentSalaryHike'
def Job_involvement(x):
    if x <= 2:
        return 'Low'
    else:
        return 'High'

attrition_df['JobInvolvement'] = attrition_df['JobInvolvement'].apply(Job_involvement)
```

```python
jbs_df = attrition_df.groupby(by=["Attrition", "JobInvolvement", "SalaryHike"]).agg({
    "EmployeeId": "count",
}).sort_values(by=["JobInvolvement"], ascending=True)
jbs_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index untuk mendapatkan kolom kembali dari multi-index
jbs_df = jbs_df.reset_index()

# Membuat bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='JobInvolvement', y='EmployeeId', hue='SalaryHike', data=jbs_df)

plt.title('Number of Employees by Job Involvement and SalaryHike')
plt.ylabel('Number of Employees')
plt.xlabel('Job Involvement')
plt.legend(title='SalaryHike')
plt.show()
```

Analisis jumlah karyawan berdasarkan SalaryHike dan PerformanceRating

```python
sp_df = attrition_df.groupby(by=["Attrition", "SalaryHike", "PerformanceRating"]).agg({
    "EmployeeId": "count",
}).sort_values(by=["SalaryHike"], ascending=True)
sp_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index untuk mendapatkan kolom kembali dari multi-index
sp_df = sp_df.reset_index()

# Membuat bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='PerformanceRating', y='EmployeeId', hue='SalaryHike', data=sp_df)

plt.title('Number of Employees by Performance rating and SalaryHike')
plt.ylabel('Number of Employees')
plt.xlabel('Performance Rating')
plt.legend(title='SalaryHike')
plt.show()
```

Analisis jumlah karyawan berdasarkan job level dan salaryHike

```python
# Membuat kolom baru 'SalaryHike' berdasarkan 'PercentSalaryHike'
def Job_level(x):
    if x <= 2:
        return 'Low'
    else:
        return 'High'

attrition_df['JobLevel'] = attrition_df['JobLevel'].apply(Job_level)
```

```python
js_df = attrition_df.groupby(by=["Attrition", "JobLevel",'SalaryHike']).agg({
    "EmployeeId": "count",
}).sort_values(by=["JobLevel"], ascending=True)
js_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset index untuk mendapatkan kolom kembali dari multi-index
js_df = js_df.reset_index()

# Membuat bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x='JobLevel', y='EmployeeId', hue='SalaryHike', data=js_df)

plt.title('Number of Employees by Job Level and SalaryHike')
plt.ylabel('Number of Employees')
plt.xlabel('Job Level')
plt.legend(title='SalaryHike')
plt.show()
```

Analisis jumlah karyawan berdasar joblevel, overtime, dan performanceRating

```python
jos_df = attrition_df.groupby(by=["Attrition", "JobLevel",'OverTime','SalaryHike']).agg({
    "EmployeeId": "count",
}).sort_values(by=["JobLevel"], ascending=True)
jos_df
```

```python
# Membuat pivot table untuk heatmap
heatmap_df = jos_df.pivot_table(index=['JobLevel', 'OverTime'], columns='SalaryHike', values='EmployeeId', aggfunc='sum')

# Membuat heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_df, annot=True, fmt="d", cmap="YlGnBu")

plt.title('Heatmap of Employee Count by Job Level, OverTime, and SalaryHike')
plt.ylabel('Job Level & OverTime')
plt.xlabel('SalaryHike')
plt.show()
```
